### Conjunto de dados do Dado Real 

```python
planta_polinizador_bruto = pd.read_csv("plant_pollinator_database.csv").iloc[:, 1:]
planta_polinizador_bruto
```

```python
planta_polinizador_bruto = pd.read_csv("plant_pollinator_database.csv").iloc[:, 1:]

planta_polinizador_bruto['season'] = planta_polinizador_bruto['season'].fillna('')
planta_polinizador_bruto['sociality'] = planta_polinizador_bruto['sociality'].replace({'no': 0, 'yes': 1})
planta_polinizador_bruto['nectar'] = planta_polinizador_bruto['nectar'].replace({'no': 0, 'yes': 1})

color_mapping = {'red': 0, 'yellow': 1, 'blue': 2, 'green': 3, 'white': 4}
planta_polinizador_bruto['colour'] = planta_polinizador_bruto['colour'].map(color_mapping)

corolla_mapping = {'CAMPANULATE': 0, 'OPEN': 1, 'TUBULAR': 2}
feeding_mapping = {'polylectic': 0, 'parasitic': 1, 'oligolectic': 2}

planta_polinizador_bruto['corolla'] = planta_polinizador_bruto['corolla'].map(corolla_mapping)
planta_polinizador_bruto['feeding'] = planta_polinizador_bruto['feeding'].map(feeding_mapping)

colunas_para_imputar = ['diameter', 'corolla', 'colour', 'nectar', 'tongue', 'body', 'sociality', 'feeding']
dados_para_imputar = planta_polinizador_bruto[colunas_para_imputar]

imputador = IterativeImputer(random_state=42)
dados_imputados = imputador.fit_transform(dados_para_imputar)

planta_polinizador_bruto[colunas_para_imputar] = dados_imputados

planta_polinizador_bruto
```

```python
planta_polinizador_bruto.isna().sum()
```

```python
traos_plantas_bruto = planta_polinizador_bruto.iloc[:, :11]
traos_plantas_bruto.columns.values[0] = "planta"
traos_plantas_bruto['planta'] = traos_plantas_bruto['planta'].astype('category')
traos_plantas_bruto
```

```python
traos_insetos_bruto = planta_polinizador_bruto.iloc[:, 11:]
traos_insetos_bruto.columns.values[0] = "inseto"
traos_insetos_bruto['inseto'] = traos_insetos_bruto['inseto'].astype('category')
traos_insetos_bruto
```

```python
matriz_interacao = pd.crosstab(index=traos_plantas_bruto['planta'], columns=traos_insetos_bruto['inseto'])
matriz_interacao
```

```python
dados = createCommunity(traos_plantas_bruto, traos_insetos_bruto, matriz_interacao)
dados
```
